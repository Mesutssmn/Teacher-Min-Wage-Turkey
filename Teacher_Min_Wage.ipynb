```python
df["GDP"]=gdp["GDP"]
```

```python
df["td"]=df["teacher"]/df["dlr"]
```

```python
df["md"]=df["min_salary"]/df["dlr"]
```

```python
df.tail()
```

```python
df["GDP"]=df["GDP"].astype("float")
```

```python
plt.style.use("dark_background")
plt.figure(figsize=(16,6))
plt.plot(df["year"],df["td"],color="red",linewidth=2)
plt.bar(df["year"],df["md"],color="magenta")
plt.legend(["Teacher $","Minimum Wage $"])
plt.title("Teacher Salary and Minimum Wage")
plt.show()
```

```python
df["td/md"] = df["td"]/df["md"]
```

```python
plt.style.use("dark_background")
plt.figure(figsize=(16,6))
plt.plot(df["year"].astype("str"),df["td/md"],color="magenta") #astype yaparak stringe çevirdiki virgüllü değerleri sildi.
plt.xticks(rotation=60) # grafikteki x değerlerini çevirdik, yticks yaparsak y eksenini yapar.
plt.legend(["Teacher $"])
plt.title("Teacher Salary/Minimum Wage")
plt.show()
```

```python
plt.style.use("dark_background")
plt.figure(figsize=(16,6))
plt.plot(df["year"].astype(str),df["GDP"]/df["td"],color="red",linewidth=2)
plt.legend(["GDP/Teacher $"])
plt.title("GDP/Teacher Salary")
plt.show()
```
